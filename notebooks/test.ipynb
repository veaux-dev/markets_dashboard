```python
def category_sheet(category):
    all_ = getattr(ta, f"{category}_all")
    rows = []
    for name in all_:
        func = getattr(ta, name)
        desc = (func.__doc__ or "").splitlines()[0]
        rows.append((name, desc))
    return pd.DataFrame(rows, columns=["Indicator", "Description"])

momentum_df = category_sheet("momentum")
```

```python
%pip install tabulate

import inspect
import tabulate 
# Candidate categories (based on repo folders)
CATEGORIES = [
    "candles", "cycles", "momentum", "overlap", "performance",
    "statistics", "trend", "utility", "volatility", "volume"
]

def extract_all_indicators():
    rows = []
    for cat in CATEGORIES:
        try:
            module = getattr(ta, cat)
        except AttributeError:
            continue

        # find all callables that look like indicators
        for name, func in inspect.getmembers(module, inspect.isfunction):
            if name.startswith("_"):  # skip internals
                continue
            doc = func.__doc__ or ""
            first_line = doc.strip().split("\n")[0] if doc else "No description"
            rows.append({
                "Category": cat.capitalize(),
                "Indicator": name,
                "Description": first_line
            })
    return pd.DataFrame(rows)

# ---- Run ----
df = extract_all_indicators()
df = df.sort_values(["Category", "Indicator"]).reset_index(drop=True)

# Export
df.to_csv("pandas_ta_all_indicators.csv", index=False, encoding="utf-8")
df.to_markdown("pandas_ta_all_indicators.md", index=False)

print(f"Extracted {len(df)} indicators across {df['Category'].nunique()} categories.")
print(df.head(15))
```
